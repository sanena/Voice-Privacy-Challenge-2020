```python
import sys


trials_file = '../corpora/test_utt_list_2/data/vctk_dev_trials_f/trials'
scores_file = '../exp/result/results-2024-01-02-15-43-10/ASV-vctk_dev_enrolls_anon-vctk_dev_trials_f_anon/scores'
trials = open(trials_file, 'r').readlines()
scores = open(scores_file, 'r').readlines()
spkrutt2target = {}
for line in trials:
  spkr, utt, target = line.strip().split()
  spkrutt2target[spkr+utt]=target
  
target_scores = []
nontarget_scores = []
for line in scores:
  spkr, utt, score = line.strip().split()
  print("{} {}".format(score, spkrutt2target[spkr+utt]))
  if (spkrutt2target[spkr+utt] == "target"):
      target_scores.append(score)
  else:
      nontarget_scores.append(score)

print(len(target_scores))
print(len(nontarget_scores))
      
```
